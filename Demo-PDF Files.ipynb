#### Extracting Metadata
We can extract some of the important useful data from any pdf. For example, we can extract the information on author of the document, its title, subject and the number of pages contained in the pdf file.

```python
from PyPDF2 import PdfReader
```

```python
import os
os.chdir("D://Files")
```

```python
def extract_pdfMeta(path):
    with open(path,'rb') as f:
        pdf = PdfReader(f)#extract text data from the interactive PDF in Python
        info = pdf.metadata
        number_of_pages = len(pdf.pages)
    print("Author: \t",info.author)
   
    print("Creator: \t",info.creator)
    
    print("Producer: \t",info.producer)
   
    print("Subject: \t",info.subject)
    
    print("title: \t",info.title)
   
    print("Number of Pages in pdf: \t",number_of_pages)
    #return info
path = 'DSA.pdf'
extract_pdfMeta(path)
```

#### Extracting Text from PDFs

```python
def extract_pdfText(path):
    with open(path, 'rb') as f:
        pdf = PdfReader(f)#extract text data from the interactive PDF in Python
      # get the 50th page
        page = pdf.pages[49]
        print(page)
        print('Page type: {}'.format(str(type(page))))
      #Extract text from the 50th page
        text = page.extract_text()
        print(text)

path = 'D://Files//DSA.pdf'
extract_pdfText(path)
```

```python
### Rotate
from PyPDF2 import PdfFileWriter
#PythonFile = open('C://Users//Nithya//Downloads//Python.pdf', 'rb')
PythonFile = open('D://Files//DSA.pdf', 'rb')
pdfReader = PyPDF2.PdfReader(PythonFile)
page = pdfReader.pages[0]
page.rotate(90)
```

```python
pdfWriter = PyPDF2.PdfWriter()
pdfWriter.add_page(page)# add pages in a PDF document in Python
resultPdfFile = open('rotatedPage.pdf', 'wb')
pdfWriter.write(resultPdfFile)
```

```python
resultPdfFile.close()
PythonFile.close()
```

```python
### merge
from PyPDF2 import PdfMerger
#Create and instance of PdfFileMerger() class
merger = PdfMerger()
#Create a list with file names
pdf_files = ['D://Files//Python.pdf', 'D://Files//DSA.pdf']
#Iterate over the list of file names
for i in pdf_files:
    #Append PDF files
    merger.append(i)
#Write out the merged PDF
merger.write("D://Nithya//merged_pages.pdf")
merger.close()
```
